# Massively parallel recordings in macaque motor cortex

Brochier and collaborator release public dataset recorded with 10-by-10 Utah electrode arrays on macaque motor cortex.

See https://www.nature.com/articles/sdata201855 for description.
Files can be downloded with gin here https://web.gin.g-node.org/INT/multielectrode_grasp


Here an example of tridesclous on one of this files : **i140703-001**.

of course the spike sorting will be done on the raw signal. this is the ns6 file in the blackrock world.



```python
# suposing the datset is downloaded here
basedir = '/media/samuel/dataspikesorting/DataSpikeSortingHD2/ThomasBrochier/multielectrode_grasp/datasets/'
filename = basedir + 'i140703-001.ns6'
```

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import tridesclous as tdc
from tridesclous import DataIO, CatalogueConstructor, Peeler
import os, shutil
```

## create a DataIO (and remove if already exists)

```python
dirname = basedir + 'tdc_i140703-001'

if os.path.exists(dirname):
    #remove is already exists
    shutil.rmtree(dirname)
    
dataio = DataIO(dirname=dirname)

# feed DataIO with one file
dataio.set_data_source(type='Blackrock', filenames=[filename])
print(dataio)

# set the probe file
dataio.set_probe_file('mea_utah_96.prb')
```

```python
fig = tdc.plot_probe_geometry(dataio, chan_grp=0)
```
